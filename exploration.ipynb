```python


from src.vectorstore import get_chroma_collection, add_chunks_to_collection

print("Getting collection...")
collection = get_chroma_collection()
print(f"Collection ready: {collection.name}")

test_chunks = [
    {"id": "test_1", "text": "The Five Pillars of Islam are Shahada, Salah, Zakat, Sawm, and Hajj.", "section": "Beliefs and practices"},
    {"id": "test_2", "text": "Islam is an Abrahamic monotheistic religion.", "section": "Introduction"}
]

print("Adding chunks...")
add_chunks_to_collection(collection, test_chunks)
print("Chunks added successfully")

print("Querying...")
results = collection.query(
    query_texts=["What are the pillars of Islam?"],
    n_results=1
)
print("Results:", results["documents"])
print("Metadata:", results["metadatas"])
```
